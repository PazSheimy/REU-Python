```python
import numpy as np
a = np.array([1,2,3])
print(type(a))
print(a.shape)
print(a[0], a[1], a[2])
```

```python
#[1,2,3] are element of the first row [4,5,6] are elements of the second row, each row have 3 colonms
b=np.array([[1,2,3],[4,5,6]])
print(b.shape)
c=np.array([[1,2,3,4],[5,6,10,12], [6,8,9,1]])
print(c.shape)
```

```python
# Numpy also provides many functions to create arrays:
z = np.zeros((4,6))
o = np.ones((2,5))
f = np.full((3,3),5)
e = np.eye(3)
r = np.random.random((4,4))
print(z)
```

```python
# Array indexing
a = np.array([[1,2,3,4],[5,6,7,8],[6,10,12,17]])
print(a)
b = a[:2,1:3]
print(b)
b[0,0] = 77
print(b)
```

```python
# Boolean arrays indexing
a = np.array([[1,2],[3,4],[5,6]])
bool_ind = (a>2)
print(bool_ind)
print(a.dtype)
print(a)
```

```python
import matplotlib.pyplot as plt 
plt.plot([1,2,3,4])
plt.ylabel('Whole numbers')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.ylabel('f(x)')
plt.xlabel('Whole numbers')
plt.show()
```

```python
#numpy array  can be plot
t = np.arange(0.,0.5,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
def f(t):
    return np.exp(-t)* np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2)), 'r--'
```
